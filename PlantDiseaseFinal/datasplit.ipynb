```python
import os
import shutil

# Paths to train, valid, and test directories
train_dir = r"D:\leaf\train"
valid_dir = r"D:\leaf\valid"
test_dir = r"D:\leaf\test"
merged_dir = r"D:\leaf\all_data"

# Create a directory to hold all data
os.makedirs(merged_dir, exist_ok=True)

# Function to move files to a single directory
def move_files(src_dir, dest_dir):
    for class_folder in os.listdir(src_dir):
        src_folder_path = os.path.join(src_dir, class_folder)
        dest_folder_path = os.path.join(dest_dir, class_folder)

        if not os.path.exists(dest_folder_path):
            os.makedirs(dest_folder_path)

        for file_name in os.listdir(src_folder_path):
            src_file_path = os.path.join(src_folder_path, file_name)
            dest_file_path = os.path.join(dest_folder_path, file_name)
            shutil.move(src_file_path, dest_file_path)

# Move files from train, valid, and test to all_data
move_files(train_dir, merged_dir)
move_files(valid_dir, merged_dir)
move_files(test_dir, merged_dir)

print("All data has been merged.")
```
